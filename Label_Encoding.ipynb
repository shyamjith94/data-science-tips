```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("bigmart.csv")
```

```python
data.head()
```

```python
data = data.drop("Item_Identifier", axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["Outlet_Size"].value_counts()
```

```python
data["Outlet_Size"].fillna("Medium", inplace=True)
```

```python
data.isnull().sum()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data["Item_Fat_Content_en"] =  le.fit_transform(data["Item_Fat_Content"])
data["Item_Type_en"] = le.fit_transform(data["Item_Type"])
data["Outlet_Size_en"] = le.fit_transform(data["Outlet_Size"])
data["Outlet_Type_en"] = le.fit_transform(data["Outlet_Type"])
```
