```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("employee.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

# One Hot Encoding

```python
x = pd.get_dummies(data["Department"])
x
```

```python
data = pd.concat([data, x], axis=1) 
```

```python
data.head()
```

```python
data = data.drop(data["Department"], axis=1, inplace=True)
```
